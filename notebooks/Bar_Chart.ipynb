## Kitabxanaların və datanın import edilməsi.

### Kitabxanalar: Pandas, Matplotlib, Seaborn

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('white') # sebaron vizuallaşdırma üçün temanın seçimi.
```

### Datanın import edilməsi. (csv faylı)

```python
df = pd.read_csv(r"C:\Users\User\Desktop\GitHub\DataVisualization\Bar_chart\data\bar_chart_data.csv")
```

## Data ilə tanış olaq.

```python
df
```

```python
df.describe().T
```

## Vizualın hazırlanması.

```python
plt.figure(figsize=(7,4))
plt.bar(x=df['Brand'],
        height=df['Cars Listings'],
        width=0.7,
        color = "teal")
plt.title("Car listings bar chart", fontsize = 15, weight = "bold")
plt.ylabel("Number of car listings", fontsize = 12, weight = "bold")
plt.xticks(rotation = 45, fontsize = 10, weight = "bold")
plt.yticks(fontsize = 10, weight = "bold")
plt.savefig(r"C:\Users\User\Desktop\GitHub\DataVisualization\Bar_chart\visuals\bar_chart.png", dpi=300, bbox_inches='tight')
plt.show()
```
